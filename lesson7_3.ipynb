```python
student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
student2 = {"國文":90, "英文":75, "數學":98, "歷史":65, "探索":75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize = (10, 8))
axes = figure.add_subplot(1, 1, 1)
#axes.plot(subjects, score1)
axes.set_title("小明和大明分數比較表")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects, score1, "ro-.", label = "小明") # "ro-." 紅色+圓點+虛線
axes.plot(subjects, score2, "go--", label = "大明") # "go--" 紅色+圓點+虛線
axes.legend() # 圖例
# axes.grid() # 圖表線條
axes.grid(axis= "y") # 圖表線條

# 加入標註，xy 是要標註的位置， xytext 是文字顯示的位置，textcoords 指定座標系統
axes.annotate(f"{student1["國文"]}", xy=(subjects[0], score1[0]), xytext=(subjects[0], score1[0]+5),
              arrowprops = dict(facecolor = "black", shrink = 0.05)) # arrowprops 讓標註有箭頭
print(subjects[0])


plt.show()
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize = (10, 8))
axes = figure.add_subplot(1, 1, 1)
#axes.plot(subjects, score1)
axes.set_title("小明和大明分數比較表")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects, score1, "ro-.", label = "小明") # "ro-." 紅色+圓點+虛線
axes.plot(subjects, score2, "gx--", label = "大明") # "go--" 紅色+x點+虛線
axes.legend() # 圖例
# axes.grid() # 圖表線條
axes.grid(axis= "y") # 圖表線條

# 加入標註，xy 是要標註的位置， xytext 是文字顯示的位置，textcoords 指定座標系統
# axes.annotate(f"{student1["國文"]}", xy=(subjects[0], score1[0]), xytext=(subjects[0], score1[0]+5),
#               arrowprops = dict(facecolor = "black", shrink = 0.05)) # arrowprops 讓標註有箭頭
# print(subjects[0])


# 標註每一個點的值
for i, (s, v1, v2) in enumerate(zip(subjects, score1, score2)):
    if s == "數學" and v1 == 92:
        axes.annotate(f"{v1}", xy = (i+0.1, v1), xytext = (i+0.4, v1),  # 由右至左
                      arrowprops = dict(facecolor = "red", 
                                        shrink = 0.05, 
                                        headlength = 8))
        axes.annotate(f"{v1}", xy = (s, v1-0.5), xytext = (s, v1-4),  # 由下至上
                      arrowprops = dict(facecolor = "red", 
                                        shrink = 0.05, 
                                        headlength = 8))
        axes.annotate(f"{v1}", xy = (i-0.1, v1), xytext = (i-0.5, v1),  # 由左至右
                      arrowprops = dict(facecolor = "red", 
                                        shrink = 0.05, 
                                        headlength = 8))
        axes.annotate(f"{v1}", xy = (s, v1+0.5), xytext = (s, v1+4),  # 由上至下
                      arrowprops = dict(facecolor = "red", 
                                        shrink = 0.05, 
                                        headlength = 8))
        axes.annotate(f"{v2}", xy = (s, v2+0.5))
    else:
        axes.annotate(f"{v1}", xy = (s, v1+1)) # y 座標+1 上移
        axes.annotate(f"{v2}", xy = (s, v2+1))

plt.show()
```
