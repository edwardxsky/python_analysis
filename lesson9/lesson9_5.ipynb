```python
# format

def rain_condition(v):
    if v < 1.75:
        return "乾燥"
    elif v < 2.75:
        return "有雨"
    else:
        return "大雨"

weather_df.style\
.set_caption("Weather Conditions")\
.background_gradient(axis=None, vmin=1, vmax=3, cmap="BuPu", subset=["Tokyo"])\
.format(rain_condition)
```

```python
# format_index

def rain_condition(v):
    if v < 1.75:
        return "乾燥"
    elif v < 2.75:
        return "有雨"
    else:
        return "大雨"

def change_index(t):
    return t.strftime("%Y年%m月%d日")


weather_df.style\
.set_caption("Weather Conditions")\
.background_gradient(axis=None, vmin=1, vmax=3, cmap="BuPu", subset=["Tokyo"])\
.format(rain_condition)\
.format_index(change_index)
```
